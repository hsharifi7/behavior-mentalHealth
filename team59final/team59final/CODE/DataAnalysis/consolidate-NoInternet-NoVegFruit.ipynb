```python
conn = sqlite3.connect("brfss.sqlite")
```

```python
query = "select `x.state`   as state , \
                MENTHLTH    as mentalhealth ,\
                `x.DRNKWEK` as drink, \
                `x.RFSMOK3` as smoke, \
                `x.EDUCAG`  as education, \
                `INTERNET`  as internet, \
                `EXERANY2`  as workout, \
                `x.FRUTSU1` as fruit, \
                `x.VEGESU1` as vegetable \
        from BRFSS_2017"                
        
    
brfssDF2017 = pd.read_sql_query(query, conn)
brfssDF2017['year']=2017
brfssDF2017
```

```python
(brfssDF2017.vegetable.isnull()).sum()
```

```python
brfssDF2017.loc[(brfssDF2017.mentalhealth == 88) , 'mentalhealth'] = 0 
brfssDF2017.loc[(brfssDF2017.mentalhealth.notnull()) & \
                (brfssDF2017.mentalhealth != 0) & \
                (brfssDF2017.mentalhealth != 77) & \
                (brfssDF2017.mentalhealth != 99), 'mentalhealth'] = 1  

brfssDF2017.loc[(brfssDF2017.drink == 0) , 'drink'] = 0 
brfssDF2017.loc[(brfssDF2017.drink != 0) & \
                (brfssDF2017.drink != 99900) , 'drink'] = 1 

brfssDF2017.loc[(brfssDF2017.smoke == 1) , 'smoke'] = 0 
brfssDF2017.loc[(brfssDF2017.smoke == 2) , 'smoke'] = 1 

brfssDF2017.loc[(brfssDF2017.education == 1) , 'education'] = 0 
brfssDF2017.loc[(brfssDF2017.education == 2) | \
                (brfssDF2017.education == 3) | \
                (brfssDF2017.education == 4), 'education'] = 1 

brfssDF2017.loc[(brfssDF2017.internet == 2) , 'internet'] = 0 
brfssDF2017.loc[(brfssDF2017.internet == 1) , 'internet'] = 1 

brfssDF2017.loc[(brfssDF2017.workout == 2) , 'workout'] = 0 
brfssDF2017.loc[(brfssDF2017.workout == 1) , 'workout'] = 1 

brfssDF2017.loc[(brfssDF2017.fruit == 0) , 'fruit'] = 0 
brfssDF2017.loc[(brfssDF2017.fruit.notnull()) & \
                (brfssDF2017.fruit != 0) , 'fruit'] = 1  

brfssDF2017.loc[(brfssDF2017.vegetable == 0) , 'vegetable'] = 0 
brfssDF2017.loc[(brfssDF2017.vegetable.notnull()) & \
                (brfssDF2017.vegetable != 0) , 'vegetable'] = 1  


brfssDF2017
```
