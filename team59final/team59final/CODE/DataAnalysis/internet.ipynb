```python
query = "select `x.state` as state , count(*) as have_mentalhealth from BRFSS_2017 \
where MENTHLTH is not null and MENTHLTH not in (88, 77, 99) \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([statesDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as have_no_mentalhealth from BRFSS_2017 \
where MENTHLTH = 88   \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as no_internet from BRFSS_2017 \
where `INTERNET`= 2 \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as internet from BRFSS_2017 \
where `INTERNET` = 1  \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)



#df.head()
#temp.head()
brfssDF2017 = temp
brfssDF2017.head()
```

```python

brfssDF = pd.concat([brfssDF2017, brfssDF2016, brfssDF2015, brfssDF2014, brfssDF2013, brfssDF2012, 
                     brfssDF2011, brfssDF2010, brfssDF2009, brfssDF2008, brfssDF2007])
brfssDF = brfssDF.dropna() 
brfssDF = brfssDF.reset_index()
brfssDF['FIPS'] = brfssDF['FIPS'].astype(int) 
brfssDF
```

```python
brfssDF = brfssDF.drop(brfssDF.columns[[0]], axis=1)
brfssDF.to_csv("raza.csv", index=False)
```

```python
temp = brfssDF
temp['p_have_mentalhealth'] = temp['have_mentalhealth']/(temp['have_mentalhealth']+temp['have_no_mentalhealth'])
temp['p_have_no_mentalhealth'] =  temp['have_no_mentalhealth']/(temp['have_mentalhealth']+temp['have_no_mentalhealth'])

temp['p_internet'] = temp['internet']/(temp['internet']+temp['no_internet'])
temp['p_no_internet'] =  temp['no_internet']/(temp['no_internet']+temp['internet'])


temp.to_csv("pdata_raza.csv", index=False)
```

```python
grouped = brfssDF.groupby(['year'])['have_mentalhealth', 'have_no_mentalhealth', 'no_internet', 'internet'].apply(lambda x: x.astype(int).sum())
grouped.reset_index()
```

```python

grouped['p_have_mentalhealth'] = grouped['have_mentalhealth']/(grouped['have_mentalhealth']+grouped['have_no_mentalhealth'])
grouped['p_have_no_mentalhealth'] =  grouped['have_no_mentalhealth']/(grouped['have_mentalhealth']+grouped['have_no_mentalhealth'])


grouped['p_internet'] = grouped['internet']/(grouped['internet']+grouped['no_internet'])
grouped['p_no_internet'] =  grouped['no_internet']/(grouped['internet']+grouped['no_internet'])

grouped = grouped.drop(grouped.columns[[0,1,2,3]], axis=1).reset_index()
grouped

```

```python
#dropping those years that we don't have data for

grouped.loc[grouped['p_no_internet'] == 0.500000, 'p_no_internet'] = float('nan') 
grouped.loc[grouped['p_internet'] == 0.500000, 'p_internet'] = float('nan') 
grouped = grouped.dropna()

grouped
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(grouped['year'],grouped['p_have_mentalhealth'])
start, end = ax.get_xlim()
ax.set_xlim(2006, 2018)

ax.set_ylim(0, 1.4)

ax.plot(grouped['year'],grouped['p_internet'])

plt.ylabel('Poportion of survey pupoluation')
plt.xlabel('Survey years')

plt.legend(loc='upper left', labels=['Have mental health', 'Use internet'])

plt.savefig('Reza.pdf')
```
