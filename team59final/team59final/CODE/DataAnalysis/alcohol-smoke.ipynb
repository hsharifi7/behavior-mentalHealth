```python
statesDF = pd.read_csv('states.csv') 
statesDF.head()
```

```python
conn = sqlite3.connect("brfss.sqlite")
```

```python
query = "select `x.state` as state , count(*) as have_mentalhealth from BRFSS_2017 \
where MENTHLTH is not null and MENTHLTH not in (88, 77, 99) \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([statesDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as have_no_mentalhealth from BRFSS_2017 \
where MENTHLTH = 88   \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as have_no_alcohol from BRFSS_2017 \
where `x.DRNKWEK`=0 \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as have_alcohol from BRFSS_2017 \
where `x.DRNKWEK` not in(0, 99900)  \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as no_smoker from BRFSS_2017 \
where `x.RFSMOK3`=1 \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)

resutlsDF = temp

query = "select `x.state` as state , count(*) as smoker from BRFSS_2017 \
where `x.RFSMOK3`=2 \
group by `x.state`;"

df = pd.read_sql_query(query, conn)
temp = pd.concat([resutlsDF, df], axis=1, sort=False).drop(['state'], axis=1)


#df.head()
#temp.head()
brfssDF2017 = temp
brfssDF2017.head()
```
