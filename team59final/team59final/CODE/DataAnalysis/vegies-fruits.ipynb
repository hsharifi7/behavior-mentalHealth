```python
temp = brfssDF
temp['p_have_mentalhealth'] = temp['have_mentalhealth']/(temp['have_mentalhealth']+temp['have_no_mentalhealth'])
temp['p_have_no_mentalhealth'] =  temp['have_no_mentalhealth']/(temp['have_mentalhealth']+temp['have_no_mentalhealth'])

temp['p_have_fruit'] = temp['have_fruit']/(temp['have_fruit']+temp['have_no_fruit'])
temp['p_have_no_fruit'] =  temp['have_no_fruit']/(temp['have_fruit']+temp['have_no_fruit'])

temp['p_have_vegetables'] = temp['have_vegetables']/(temp['have_no_vegetables']+temp['have_vegetables'])
temp['p_have_no_vegetables'] =  temp['have_no_vegetables']/(temp['have_no_vegetables']+temp['have_vegetables'])

temp.to_csv("pdata_karthik.csv", index=False)
```

```python
grouped = brfssDF.groupby(['year'])['have_mentalhealth', 'have_no_mentalhealth', 'have_no_fruit', 'have_fruit', 'have_no_vegetables', 'have_vegetables'].apply(lambda x: x.astype(int).sum())
grouped.reset_index()
```

```python

grouped['p_have_mentalhealth'] = grouped['have_mentalhealth']/(grouped['have_mentalhealth']+grouped['have_no_mentalhealth'])
grouped['p_have_no_mentalhealth'] =  grouped['have_no_mentalhealth']/(grouped['have_mentalhealth']+grouped['have_no_mentalhealth'])

grouped['p_have_fruit'] = grouped['have_fruit']/(grouped['have_fruit']+grouped['have_no_fruit'])
grouped['p_have_no_fruit'] =  grouped['have_no_fruit']/(grouped['have_fruit']+grouped['have_no_fruit'])

grouped['p_have_vegetables'] = grouped['have_vegetables']/(grouped['have_no_vegetables']+grouped['have_vegetables'])
grouped['p_have_no_vegetables'] =  grouped['have_no_vegetables']/(grouped['have_no_vegetables']+grouped['have_vegetables'])

grouped = grouped.drop(grouped.columns[[0,1,2,3,4,5]], axis=1).reset_index()
grouped

```

```python
#dropping those years that we don't have data for

grouped.loc[grouped['p_have_fruit'] == 0.500000, 'p_have_fruit'] = float('nan') 
grouped.loc[grouped['p_have_no_fruit'] == 0.500000, 'p_have_no_fruit'] = float('nan') 
grouped.loc[grouped['p_have_vegetables'] == 0.500000, 'p_have_vegetables'] = float('nan') 
grouped.loc[grouped['p_have_no_vegetables'] == 0.500000, 'p_have_no_vegetables'] = float('nan') 
grouped = grouped.dropna()

grouped
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(grouped['year'],grouped['p_have_mentalhealth'])
start, end = ax.get_xlim()
ax.set_xlim(2006, 2018)

ax.set_ylim(0, 1.6)

ax.plot(grouped['year'],grouped['p_have_fruit'])

ax.plot(grouped['year'],grouped['p_have_vegetables'])


plt.ylabel('Poportion of survey pupoluation')
plt.xlabel('Survey years')

plt.legend(loc='upper left', labels=['Have mental health', 'Eat fruits', 'Eat vegetables'])


plt.savefig('karthik.pdf')
```
